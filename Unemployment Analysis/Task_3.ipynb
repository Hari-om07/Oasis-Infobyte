```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Oasis Infobyte/Unemployment Analysis/Unemployment in India.csv')
data.head()
```

```python
print("Missing values in Data:")
print(data.isnull().sum())
```

```python
#Drop empty rows
data1 = data.dropna()
```

```python
print("Missing values in Data:")
print(data1.isnull().sum())
```

```python
data1["Region"].value_counts()
```

```python
data1["Area"].value_counts()
```

```python
Insight1 = data1["Area"].value_counts()
```

```python
def format_y_ticks(value, _):
    return "{:.1f}".format(value/10)

plt.figure(figsize = (8,6))
plt.bar(data1["Area"], data1[" Estimated Unemployment Rate (%)"], color = "skyblue")
plt.title("Estimated Unemployment Rate (%) by Area")
plt.xlabel("Area")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_y_ticks))
plt.tight_layout()
plt.show()
```

```python
Insight2 = data1["Region"].value_counts()
```

```python
def format_y_ticks(value , _):
    return "{:.1f}".format(value/10)
plt.figure(figsize=(10, 6))
plt.bar(data1['Region'], data1[' Estimated Unemployment Rate (%)'], color='skyblue')
plt.title('Estimated Unemployment Rate (%) by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_y_ticks))
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
#Insight3
plt.figure(figsize = (8,6))
plt.scatter(data1[' Estimated Employed'], data1[' Estimated Labour Participation Rate (%)'], color = "orange", alpha = 0.7)
plt.title("Employed VS Labour Participation")
plt.xlabel("Estimated Employed")
plt.ylabel("Estimated Labour Participation Rate (%)")
plt.tight_layout()
plt.show()
```

```python
data1[" Frequency"] = data1[" Frequency"].astype(str)
Insight4 = data1[" Frequency"].value_counts()
```

```python
plt.figure(figsize = (8,6))
Insight4.plot(kind = 'bar', color = 'purple')
plt.title("Frequency of Unemployment Rate Reporting")
plt.xlabel("Frequency")
plt.ylabel("Counts")
plt.tight_layout()
plt.show()
```
